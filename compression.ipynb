```python
import cv2
import matplotlib.pyplot as plt
import os
# Load the image
image_path = '/content/download (3).jfif'
image = cv2.imread('/content/download (3).jfif')

# Convert image to BGR for OpenCV
image_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
```

```python
# Save the image with lossless compression (PNG)
cv2.imwrite('image_lossless.png', image_bgr)

# Save the image with lossy compression (JPEG)
cv2.imwrite('image_lossy.jpg', image_bgr, [int(cv2.IMWRITE_JPEG_QUALITY), 90])

```

```python
# Load the compressed images
image_lossless = cv2.imread('image_lossless.png')
image_lossy = cv2.imread('image_lossy.jpg')

# Convert images from BGR to RGB
image_bgr = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
image_lossless = cv2.cvtColor(image_lossless, cv2.COLOR_BGR2RGB)
image_lossy = cv2.cvtColor(image_lossy, cv2.COLOR_BGR2RGB)

# Plot the images
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image_bgr)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Lossless Compression (PNG)')
plt.imshow(image_lossless)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Lossy Compression (JPEG)')
plt.imshow(image_lossy)
plt.axis('off')

plt.show()
```

```python
# Calculate dimensions
original_dim = image_bgr.shape
lossless_dim = image_lossless.shape
lossy_dim = image_lossy.shape

# Calculate file sizes
original_size = os.path.getsize('/content/download (3).jfif')  # Use the actual path to the original image
png_size = os.path.getsize('image_lossless.png')
jpeg_size = os.path.getsize('image_lossy.jpg')

# Function to calculate memory usage in bytes
def calculate_memory_usage(image):
    if image is None:
        return 0
    height, width, channels = image.shape
    return height * width * channels  # Each channel is 1 byte for standard images

# Calculate memory usage
memory_original = calculate_memory_usage(image_bgr)
memory_lossless = calculate_memory_usage(image_lossless)
memory_lossy = calculate_memory_usage(image_lossy)

# Print dimensions, file sizes, and memory usage
print(f"Original Image - Dimensions: {original_dim[1]}x{original_dim[0]}, Size: {original_size / 1024:.2f} KB, Memory usage: {memory_original / (1024 * 1024):.2f} MB")
print(f"Lossless Compression (PNG) - Dimensions: {lossless_dim[1]}x{lossless_dim[0]}, Size: {png_size / 1024:.2f} KB, Memory usage: {memory_lossless / (1024 * 1024):.2f} MB")
print(f"Lossy Compression (JPEG) - Dimensions: {lossy_dim[1]}x{lossy_dim[0]}, Size: {jpeg_size / 1024:.2f} KB, Memory usage: {memory_lossy / (1024 * 1024):.2f} MB")
```
