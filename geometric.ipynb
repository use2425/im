```python
import cv2
import numpy as np

# Step 3: Create a black and white image
image = np.zeros((450, 650, 3), dtype=np.uint8)  # black image of size 600x400

# Step 4: Draw a line
cv2.line(image, (50, 50), (550, 50), (255, 0, 0), 8)  # Draw a blue line from (50,50) to (550,50)

# Step 5: Draw an arrowed line
cv2.arrowedLine(image, (50, 100), (550, 100), (0, 255, 0), 5)  # Draw a green arrowed line

# Step 6: Draw a rectangle
cv2.rectangle(image, (50, 150), (550, 350), (0, 0, 255), 3)  # Draw a red rectangle

# Step 7: Draw a circle
cv2.circle(image, (300, 350), 50, (255, 255, 0), -1)  # Draw a filled yellow circle

# Step 8: Draw text
cv2.putText(image, 'TARAAAAAA', (50, 380), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)

# Display the image
from matplotlib import pyplot as plt
plt.imshow(image)
plt.axis('off')  # Hide the axis
plt.show()
```
