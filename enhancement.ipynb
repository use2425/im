```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2https://accounts.google.com/SignOutOptions?hl=en-GB&continue=https://colab.research.google.com/%3Fauthuser%3D0&ec=GBRAqQM

# Load an image from your drive
image_path = '/content/download (1).jfif'
image = Image.open(image_path)

# Convert image to grayscale
gray_image = image.convert('L')

# Display the original and grayscale images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image)

plt.subplot(1, 2, 2)
plt.title('Grayscale Image')
plt.imshow(gray_image, cmap='gray')

plt.show()
```

```python
# Step 4: Adjust brightness and contrast
def adjust_brightness_contrast(image, brightness=0, contrast=0):
    # Convert PIL Image to NumPy array
    image_np = np.array(image)
    new_image = cv2.convertScaleAbs(image_np, alpha=contrast, beta=brightness)  # Pass the NumPy array
    return new_image

adjusted_image = adjust_brightness_contrast(image, brightness=40, contrast=1.5)

# Display the adjusted image
plt.figure(figsize=(5, 5))
plt.title('Brightness & Contrast Adjusted Image')
plt.imshow(adjusted_image)
plt.show()

```

```python
# Convert image to numpy array
img_array = np.array(image)

# Find the digital negative
negative_image = 255 - img_array

# Convert back to image
negative_image = Image.fromarray(negative_image)

# Display the digital negative
plt.figure(figsize=(5, 5))
plt.title('Digital Negative Image')
plt.imshow(negative_image)
plt.show()

```

```python
# Get RGB values of each pixel
rgb_values = np.array(image)

# Display the RGB values of the first 5 pixels
print(rgb_values[:5, :5, :])
```

```python
# Create new color values by subtracting from 255
new_rgb_values = 255 - rgb_values
# Set the new pixel values (digital negative)
new_image = Image.fromarray(new_rgb_values)

# Display the new image
plt.figure(figsize=(5, 5))
plt.title('New Image with Modified Pixel Values')
plt.imshow(new_image)
plt.show()
```
