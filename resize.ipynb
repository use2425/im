```python
#Name:SK Tarannum
#Roll.no:35
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Replace 'path_to_your_image' with the path to your image file
image_path = '/content/download.jfif'

# Load the color image
color_image = cv2.imread(image_path)

# Convert the color image to grayscale
gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

# Convert the grayscale image to binary using a threshold
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

# Resize the images
resized_color_image = cv2.resize(color_image, (200, 200))  # Resize to 100x100 pixels
resized_gray_image = cv2.resize(gray_image, (500, 500))    # Resize to 100x100 pixels
resized_binary_image = cv2.resize(binary_image, (100, 100))  # Resize to 100x100 pixels

#Copy the resized images into other variables
copied_color_image = resized_color_image.copy()
copied_gray_image = resized_gray_image.copy()
copied_binary_image = resized_binary_image.copy()


# Display the copied resized images and original images:
print("Original Color Image")
cv2_imshow(color_image)
print("Copied Resized Color Image:")
cv2_imshow(copied_color_image)
print("Original GrayScale Image")
cv2_imshow(gray_image)
print("Copied Resized Grayscale Image:")
cv2_imshow(copied_gray_image)
print("Original Binary Image")
cv2_imshow(binary_image)
print("Copied Resized Binary Image:")
cv2_imshow(copied_binary_image)
```

```python
# Check the size and type of the new image
print("Resized Color Image Shape:", resized_color_image.shape)
print("Resized Greyscale Image Shape:", resized_gray_image.shape)
print("Resized Binary  Image Shape:", resized_binary_image.shape)
print("Resized Image Type:", type(resized_color_image))
```
